### Import libraries

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

import matplotlib
font = {'weight' : 'normal',
        'size'   : 12}

matplotlib.rc('font', **font)
```

### Load data

```python
data = pd.read_csv("Figure2-3-S2-double_mutant_validation.csv")
data
```

### Plot

```python
exp_average = data[['Expression_rep1', 'Expression_rep2']].mean(axis=1)
exp_range = np.abs(data['Expression_rep1'] - data['Expression_rep2'])

plt.figure(figsize=(16,6))

plt.bar(range(len(exp_average)), exp_average, yerr=exp_range, capsize=2, edgecolor='k',  color=data["Color"])
plt.xticks(range(len(exp_average)), data["Mutant"], rotation=90)

plt.ylim([0,2])
plt.xlim([-0.7, 22.7])
plt.grid(axis='y', linestyle=":")

plt.plot([-1,30],[1,1], 'k--', linewidth=1.25)

plt.xlabel("ACE2 variant")
plt.ylabel("Relataive expression")
plt.show()
```
